```python
min_votes = df['vote_count'].quantile(0.9)
min_votes
```

```python
df = df.loc[df['vote_count'] >= min_votes]
df.shape
```

```python
def weighted_rating(df, min_votes=min_votes, vote_avg=vote_avg):
    v = df['vote_count']
    R = df['vote_average']
    
    return (v/(v+min_votes) * R) + (min_votes/(min_votes+v) * vote_avg)
```

```python
df['score'] = df.apply(weighted_rating, axis=1)
df.head()
```

# Finding similarity on basis of overview

```python
#  TF = no of repetitions of the word in sentence / no of total words in sentence
# IDF = no of sentences / no of sentences containing the word
```

```python
from sklearn.feature_extraction.text import TfidfVectorizer

# Define a TF-IDF vectorizer object. Remove all stopwords
tfidf = TfidfVectorizer(stop_words = 'english')

# Replace NaN/null values with empty string
movies['overview'] = movies['overview'].fillna('')

# Construct the TF-IDF matrix by fitting and transforming the data
tfidf_matrix = tfidf.fit_transform(movies['overview'])

tfidf_matrix.shape
```

```python
from sklearn.metrics.pairwise import linear_kernel

# Compute cosine similarity matrix
cosine_sim = linear_kernel(tfidf_matrix, tfidf_matrix)
```

```python
indices = pd.Series(movies.index, index=movies['title']).drop_duplicates()
indices
```

```python
def get_recommendations(title, cosine_sim=cosine_sim):
    
    # Get the index of the movie
    idx = indices[title]
    
    # Get pairwise similarity scores of all movies with the given movie
    sim_scores = list(enumerate(cosine_sim[idx]))
    
    # Sort the movies based on similarity scores
    sim_scores = sorted(sim_scores, key=lambda x : x[1], reverse=True)
    
    # Get the scores of the top 10 similar movies
    sim_scores = sim_scores[1:11]
    
    # Get the movie indices
    movie_indices = [i[0] for i in sim_scores]
    
    # Return top 10 similar movies
    return movies['title'].iloc[movie_indices]
```

```python
get_recommendations('The Dark Knight Rises')
```

# Based on Genres, Keywords, Cast & Crew

```python
# Get the diretor's name

import json

def get_director(x):
    js = json.loads(x)
    length = len(js)
    
    for i in range(length):
        c = js[i]
        if c['job'] == 'Director':
            return c['name']
    return np.nan
```

```python
# Returns first 3 elements from the list

def get_list(x):
    js = json.loads(x)
    length = len(js)
    
    names = []
    
    if isinstance(js, list):
        for i in range(length):
            names.append(js[i]['name'])
        
        if len(names) > 3:
            names = names[:3]
        return names
    
    return []
```

```python
# Define new columns for director, cast, genres and keywords
movies['director'] = movies['crew'].apply(get_director)

features = ['cast', 'genres', 'keywords']

for feature in features:
    movies[feature] = movies[feature].apply(get_list)
```

```python
movies[['title', 'cast', 'director', 'keywords', 'genres']].head(5)
```

```python
def clean_data(x):
    if isinstance(x, list):
        return [str.lower(i.replace(" ", "")) for i in x]
    else:
        if isinstance(x, str):
            return str.lower(x.replace(" ", ""))
        else:
            return ''
```

```python
features = ['cast', 'keywords', 'director', 'genres']

for feature in features:
    movies[feature] = movies[feature].apply(clean_data)
```

```python
movies[['title', 'cast', 'director', 'keywords', 'genres']].head(5)
```

# Creating MetaData

```python
def create_soup(x):
    return ' '.join(x['keywords']) + ' ' + ' '.join(x['cast']) + ' ' + (x['director']) + ' ' + ' '.join(x['genres'])

movies['soup'] = movies.apply(create_soup, axis=1)
```

```python
movies['soup']
```

```python
from sklearn.feature_extraction.text import CountVectorizer

count = CountVectorizer(stop_words='english')
count_matrix = count.fit_transform(movies['soup'])
```

```python
from sklearn.metrics.pairwise import cosine_similarity

cosine_simi = cosine_similarity(count_matrix, count_matrix)
```
